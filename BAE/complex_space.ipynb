```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 실수 및 허수 부분 정의
X_real = np.linspace(-2, 2, 400)
X_imag = np.linspace(-2, 2, 400)
X_real, X_imag = np.meshgrid(X_real, X_imag)

# 지수 함수 계산 (a=2)
a = 2
X = X_real + 1j * X_imag
Y = a ** X

# 실수 및 허수 부분 분리
Y_real = np.real(Y)
Y_imag = np.imag(Y)

# 3D 그래프 그리기
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# 실수 부분 그래프
ax.plot_surface(X_real, X_imag, Y_real, cmap='viridis', alpha=0.6)

# 허수 부분 그래프
ax.plot_surface(X_real, X_imag, Y_imag, cmap='plasma', alpha=0.6)

# 축 라벨 및 제목 설정
ax.set_xlabel('Real Part of x')
ax.set_ylabel('Imaginary Part of x')
ax.set_zlabel('Value of 2^x')
plt.title('Complex Exponential Function 2^x in 3D')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the real and imaginary parts of x
X_real = np.linspace(-2, 2, 400)
X_imag = np.linspace(-2, 2, 400)
X_real, X_imag = np.meshgrid(X_real, X_imag)

# Calculate e^x
X = X_real + 1j * X_imag
Y = np.exp(X)

# Separate the result into real and imaginary parts
Y_real = np.real(Y)
Y_imag = np.imag(Y)

# Plot the 3D surface
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot real part of Y
ax.plot_surface(X_real, X_imag, Y_real, cmap='viridis', alpha=0.6)

# Plot imaginary part of Y
ax.plot_surface(X_real, X_imag, Y_imag, cmap='plasma', alpha=0.6)

ax.set_xlabel('Real Part of x')
ax.set_ylabel('Imaginary Part of x')
ax.set_zlabel('Value of e^x')

plt.title('Complex Exponential Function e^x in 3D')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range for r (radius) and theta (angle)
r = np.linspace(0.01, 2, 400)  # Avoid r=0 to prevent log(0)
theta = np.linspace(-np.pi, np.pi, 400)
r, theta = np.meshgrid(r, theta)

# Convert polar coordinates to complex numbers
X = r * np.exp(1j * theta)
Y = np.log(X)

# Separate the result into real and imaginary parts
Y_real = np.real(Y)
Y_imag = np.imag(Y)

# Plot the 3D surface
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot real part of Y
ax.plot_surface(r, theta, Y_real, cmap='viridis', alpha=0.6)

# Plot imaginary part of Y
ax.plot_surface(r, theta, Y_imag, cmap='plasma', alpha=0.6)

# Labels and title
ax.set_xlabel('Radius (r)')
ax.set_ylabel('Angle (theta)')
ax.set_zlabel('Value of ln(x)')
plt.title('Complex Logarithm Function ln(x) in Polar Coordinates')

plt.show()
```
